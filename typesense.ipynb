```python
import json

with open("books.jsonl", "r", encoding="utf-8-sig") as f:  # utf-8-sig removes BOM
    lines = f.readlines()

with open("books_clean.jsonl", "w", encoding="utf-8") as out:
    for line in lines:
        line = line.strip()
        if not line:
            continue  # ðŸš¨ remove empty lines
        obj = json.loads(line)  # will crash if invalid â†’ good
        out.write(json.dumps(obj, ensure_ascii=False) + "\n")
```
